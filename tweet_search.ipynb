```python
## Yiğit Güleç

import twint
import nest_asyncio
import pandas as pd
from IPython.display import HTML
import requests
nest_asyncio.apply()
```

```python
c = twint.Config()

# I Will limit number of tweets as 100, otherwise it will store all recent tweets.
c.Search = "request for startup"
c.Pandas = True
c.Limit = 100 ## if we ommit this part we can get limitless data
twint.run.Search(c)
```

```python
Tweets_df = twint.storage.panda.Tweets_df
```

```python
# to show our tweets in a html page format
def show_tweet(link):
    '''Display the contents of a tweet. '''
    url = 'https://publish.twitter.com/oembed?url=%s' % link
    response = requests.get(url)
    html = response.json()["html"]
    display(HTML(html))
```

```python
for i in range(1, len(Tweets_df)):
    sample_tweet_link = Tweets_df.sample(i)['link'].values[0]
    display(sample_tweet_link)
    show_tweet(sample_tweet_link)
Tweets_df.to_csv(r'all_tweets.csv')
```

```python
columns = ['username', 'name', 'tweet', 'date', 'nlikes', 'nreplies', 'nretweets','link']
# sorted tweets according to number of retweets
temp = Tweets_df.sort_values(by= ["nretweets"], ascending = False)
sort_by_retweets = temp[columns]
sort_by_retweets = sort_by_retweets.reset_index(drop = True)
sort_by_retweets.to_csv(r'all_tweets_retweet_descending.csv')
```

```python
# sorted tweets according to number of discussions
temp = Tweets_df.sort_values(by= "nreplies", ascending = False)
sort_by_discussions = temp[columns]
sort_by_discussions = sort_by_discussions.reset_index(drop = True)
sort_by_discussions.to_csv(r'all_tweets_discussion_descending.csv')
```

```python
# sorted tweets according to number of likes
temp = Tweets_df.sort_values( by = ['nlikes'] , ascending = False)
sort_by_likes = temp[columns]
sort_by_likes = sort_by_likes.reset_index(drop = True)
sort_by_likes.to_csv(r'all_tweets_like_descending.csv')
```

```python
# sorted tweets according to date
temp = Tweets_df.sort_values(by= ["date"], ascending = False)
sort_by_date = temp[columns]
sort_by_date = sort_by_date.reset_index(drop = True)
sort_by_date.to_csv(r'all_tweets_date_descending.csv')
```
